```python
# Thực hành 1: Tích phân không xác định loại 1
from sympy import Symbol, Integral
x = Symbol('x')
f = 1.0/x
unbound1 = Integral(f, (x, 0, 2))
unbound1.doit() 
```

```python
# Thực hành 2: Tích phân không xác định loại 2:
from sympy import Symbol, Integral
x = Symbol('x')
f = 1.0/x**2
unbound2 = Integral(f, (x, -1, 1))
unbound2.doit()
```

```python
# Thực hành 3: Tích phân không liên tục
from sympy import Symbol, Integral, Abs
x = Symbol('x')
f = Abs(x)
discontinous1 = Integral(f, (x, -1, 2))
discontinous1.doit() 
```

```python
# Thực hành 4: Tích phân có hàm thỏa không tồn tại trong miền xác định (thực)
from sympy import Symbol, Integral
x = Symbol('x')
f = x**(0.5)
undef1 = Integral(f, (x, -1, 1))
undef1.doit() 
```

```python
# Thực hành 5: Một máy đo đo được lượng thuốc nhuộm trong máu với số liệu cho sẵn:
from scipy import integrate
import numpy as np
y = np.array([0.0, 4.1, 8.9, 8.5, 6.7, 4.3, 2.5, 1.2, 0.2]) 
x = np.arange(0,9)
integrate.simps(y, x)
S = 2*integrate.simps(y, x)
print(S)
A = 5.5
F = A/S * 60
print(F)
```

```python
# Thực hành 6: Hãy viết các câu lệnh Python để tìm độ dài của đường cong:
# Khai báo thư viện sympy
import sympy as sp
# Khai báo biến và hàm f
x = sp.symbols('x')
f = x**(sp.Rational(3,2))
# Tính đạo hàm của f
df = sp.diff(f, x)
# Tính độ dài đường cong
L = sp.integrate(sp.sqrt(1 + df**2), (x, 0, 2))
# Hiển thị kết quả
print("f(x) =", f)
print("f'(x) =", df)
print("Độ dài đường cong L =", L)
print("Giá trị gần đúng =", sp.N(L))
```

```python
# Thực hành 7: Hãy vẽ 3 đoạn đường sắt sau khi thiết kế xong đoạn nối (với −2 ≤  ≤ 2)
import numpy as np
import matplotlib.pyplot as plt

# Tạo các khoảng x
x1 = np.linspace(-2, 0, 100)     # đoạn 1
x2 = np.linspace(0, 1, 100)      # đoạn nối
x3 = np.linspace(1, 2, 100)      # đoạn 2

# Các hàm
y1 = 0*x1
y2 = -8*x2**3 - 8*x2**4 + 3*x2**5
y3 = x3

# Vẽ đồ thị
plt.figure(figsize=(7,5))
plt.plot(x1, y1, 'b', label='Đoạn 1: y=0')
plt.plot(x2, y2, 'r', label='Đoạn nối')
plt.plot(x3, y3, 'g', label='Đoạn 2: y=x')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Thiết kế đoạn nối đường sắt")
plt.legend()
plt.grid(True)

plt.xlim(-2,2)
plt.ylim(-1,2)

plt.show()
```
