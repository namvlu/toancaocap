```python
# Thực hành 1: Lagrange
from sympy import symbols, solve
x, y, l = symbols('x y l')
# Hàm Lagrange
f1 = 1 - l*y
f2 = 2 - l*x
f3 = x*y - 5000
sol = solve((f1, f2, f3), (x, y, l))
print(sol)
```

```python
# Thực hành 2: Thủy triều
import sympy as sp
t = sp.symbols('t')
def d(t_val):
    return -2.515*sp.sin(0.503*t_val - 3.39525)
print("3h:", d(3).evalf())
print("6h:", d(6).evalf())
print("9h:", d(9).evalf())
print("12h:", d(12).evalf())
```

```python
# Thực hành 3: Hồi quy bậc 1
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
x_data = np.array([3.4, 4.6, 5.2, 8.0, 10.7])
y_data = np.array([48, 52, 58, 76, 96])
def f(x, a, b):
    return a*x + b
params, _ = curve_fit(f, x_data, y_data)
print("a, b =", params)
y_73 = f(7.3, *params)
print("Giá trị tại 7.3 =", y_73)
plt.scatter(x_data, y_data)
plt.plot(x_data, f(x_data, *params))
plt.show()
```

```python
# Thực hành 4: Hồi quy bậc 2
def f2(x, a, b, c):
    return a*x**2 + b*x + c
params, _ = curve_fit(f2, x_data, y_data)
print("a, b, c =", params)
y_73 = f2(7.3, *params)
print("Giá trị tại 7.3 =", y_73)
```

```python
# Thực hành 5: Tích phân kép
from sympy import *
x, y = symbols('x y')
f = 3*x**2 - y
tp1 = integrate(f, (y, 0, 3), (x, 0, 2))
tp2 = integrate(f, (x, 0, 2), (y, 0, 3))
print(tp1)
print(tp2)
```

```python
# Thực hành 6: Tích phân miền cong
from sympy import *
x, y = symbols('x y')
f = x*y
tp = integrate(f, (y, x, 2 - x**2), (x, -2, 1))
print(tp)
```

```python
# Thực hành 7: Lượng mưa trung bình
def luong_mua_trung_binh(data, tong_dien_tich):
    n = len(data)
    delta = tong_dien_tich / n
    return sum(data) * delta / tong_dien_tich
data = [0.08, 0.08, 0.01, 1.7, 1.74, 3.0]
print(luong_mua_trung_binh(data, 75000))
```
