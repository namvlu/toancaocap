```python
# Thực hành 1: Viết hàm đệ quy tính n!
import math
def giaithua(n):
    if (n==0):
        return 1
    else:
        return n*giaithua(n-1)
print(giaithua(3))
print(giaithua(4))
```

```python
# Thực hành 2: Viết hàm đệ quy tính dãy số Fibonacci
def fibo(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibo(n-1) + fibo(n-2)
fibo(4)
```

```python
fibo(5)
```

```python
fibo(6)
```

```python
# Thực hành 3: Chứng minh dãy có giới hạn và tìm giới hạn của dãy
def an_exp_an(n):
    if n == 1:
        return 1.0/2
    else:
        return an_exp_an(n-1)**an_exp_an(n-1)
an_exp_an(1)
```

```python
an_exp_an(10)
```

```python
# Thực hành 4: Chứng minh dãy có giới hạn và tìm giới hạn của dãy
def bai4(n):
    if n==1:
        return 4
    else:
        return 5.0/(6-bai4(n-1))
bai4(5)
```

```python
bai4(100)
```

```python
# Thực hành 5: Xét giới hạn của dãy khi biết dãy Fibonacci được định nghĩa
def tisoFibo(n):
    if n==1:
        return 1
    else:
        return 1.0 + 1.0/tisoFibo(n-1)
for i in range(1, 11):
    print (i, tisoFibo(i))
```
