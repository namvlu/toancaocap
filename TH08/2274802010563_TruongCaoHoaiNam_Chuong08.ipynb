```python
# Thực hành 8:
from sympy import Derivative, Symbol, solve
x = Symbol('x')
y = Symbol('y')
z = Symbol('z')
f = 4*x*y - 8*x - 5 + z*(x+y-50)
dx = Derivative(f, x).doit()
dy = Derivative(f, y).doit()
dz = Derivative(f, z).doit()
nghiem = solve([dx, dy, dz], (x,y,z))
print (nghiem)
```

```python
xx = nghiem[x]
yy = nghiem[y]
4*xx*yy - 8*xx - 5 
```

```python
ketqua = 4*nghiem[x]* nghiem[y] - 8* nghiem[x] - 5
print (ketqua) 
```
